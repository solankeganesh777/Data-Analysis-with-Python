```python
# Data Acquisition

import pandas as pd

path="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
df=pd.read_csv(path,header=None)
```

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#Add headers on dataframe
df.columns=headers
df.head()
```

```python
# Missing data: Drop missing Values
df.dropna(subset=["price"],axis=0)
```

```python
# Data Wrangling

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Replace '?' to "NaN"
df.replace("?",np.nan,inplace=True)
df.head()
```

```python
# Identify missing values: Evaluating for missing data 
missing_data=df.isnull()
missing_data.head()
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

```python
#Missing data: Replace by mean 

#Here, its done for 'normalized-losses column'
avg_norm_loss=df["normalized-losses"].astype("float").mean(axis=0)
df['normalized-losses'].replace(np.nan,avg_norm_loss,inplace=True)

print(avg_norm_loss)
print(df['normalized-losses'])
```

```python
#Replace by mean for column bore
avg_bore=df['bore'].astype('float').mean(axis=0)
df['bore'].replace(np.nan,avg_bore,inplace=True)
print(df['bore'])
```

```python
# Check data types
df.dtypes
```

```python
#Data Standardization
print(df.head())

df['city-L/100km']=235/df['city-mpg']

print(df.head())
```

```python
# Data Normalization

# SImple feature scailing
df['length']=df['length']/df['length'].max()

#Min-max normalization
df['width']=(df['width']-df['width'].min())/(df['width'].max()-df['width'].min())


df[['length','width']].head()
```

```python
# Data Binning
import matplotlib.pyplot as plt

# Convert data to correct format
df['horsepower']=df['horsepower'].astype('Int32',copy=True)

# Plot histogram of horsepower to see distribution of horsepower
plt.hist(df['horsepower'])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
```

```python
bins=np.linspace(min(df['horsepower']),max(df['horsepower']),4)
bins
```

```python
# Set group names for bins
group_names=['low','medium','high']
```

```python
df['horsepower-binned']=pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower','horsepower-binned']].head(20)
```

```python
# Values in each bins
df['horsepower-binned'].value_counts()
```

```python
# Plot distribution of each bin

plt.bar(group_names,df['horsepower-binned'].value_counts())
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")

plt.show()
```

```python
# Histogram to viualize distribution of bins created

plt.hist(df['horsepower'],bins=3)
plt.show()
```

```python
# Indicator variable/ Dummy variable
df.columns
```

```python
# Assign numerical values to different categories of fuel type. 
# Get indicator variables and assign it to data frame "dummy_variable_1" 

dummy_variable_1=pd.get_dummies(df['fuel-type'])
dummy_variable_1.head()
```

```python
# Change columns names for clarity
dummy_variable_1.rename(columns={'fuel-type-disel':'gas','fuel-type-disel':'disel'},inplace=True)
dummy_variable_1.head()
```

```python
# We now have the value 0 to represent "gas" and 1 to represent "diesel" in the column "fuel-type".
# We will now insert this column back into our original dataset

# Merge data frame 'df' and 'dummy_variable_1'
df=pd.concat([df,dummy_variable_1],axis=1)
```

```python
# Drop original column "fuel-type" from df
df.drop("fuel-type",axis=1,inplace=True)
```
